##Price Elasticity of Demand Analysis

Price Elasticity of Demand (PED) measures the responsiveness of the quantity demanded of a product to changes in its price. It is calculated as the percentage change in quantity demanded divided by the percentage change in price. It helps businesses understand how sensitive their customers are to price changes, which enables them to make informed pricing decisions.

```python
#Clculo do percentual de mudan em preo e quantidade
df['Price_change'] = df.groupby(['Store_ID', 'Item_ID'])['Price'].pct_change()
df['Quantity_change'] = df.groupby(['Store_ID', 'Item_ID'])['Item_Quantity'].pct_change()
```

```python
#Clculo da elasticidade do preo da demanda
df['PED'] = df['Quantity_change']/df['Price_change']
```

```python
df.replace([float('inf'), -float('inf')], float('nan'), inplace=True)
df.dropna(subset = ['PED'], inplace = True)
```

```python
import pandas as pd

def resumo_estatistico(df, coluna):
    """
    Exibe estatsticas descritivas da coluna especificada.
    """
    resumo = df[coluna].describe(percentiles=[0.25, 0.5, 0.75])
    print(resumo)


resumo_estatistico(df, 'PED')
```

O valor mdio do PED  de -0.46 o que indica que na mdia a demanda  aproximadamento inesltica. resumidamente, um acrscimo de 1% no preo resulta em 0.35% de decrscimo na quantidade demandada


```python
#definindo os threshholds de PED
elastic_threshold = 1
inelastic_threshold = -1

# segment the data based on PED
df['Segment'] = 'Unitary Elastic'
df.loc[df['PED'] > elastic_threshold, 'Segment'] = 'Highly Elastic'
df.loc[df['PED'] < inelastic_threshold, 'Segment'] = 'Inelastic'
df.loc[df['PED'] == 0, 'Segment'] = 'Zero Elasticity'
df.loc[df['PED'] < 0, 'Segment'] = 'Negative Elasticity'

# count the number of items in each segment
segment_counts = df['Segment'].value_counts()

segment_counts
```

```python
fig_segment = px.scatter(df, x='Price_change', y='Quantity_change', color='Segment',
                         title='Price Change vs Quantity Change Across Different Market Segments',
                         labels={'Price_change': 'Price Change (%)', 'Quantity_change': 'Quantity Change (%)'},
                         color_continuous_scale='Viridis', opacity=0.6)

fig_segment.update_layout(xaxis_title='Price Change (%)', yaxis_title='Quantity Change (%)',
                          template='plotly_white')

fig_segment.show()
```

O grfico representa a relao entre a variao percentual de preo e a variao percentual de quantidade em diferentes segmentos de mercado, categorizados por elasticidade. Os segmentos incluem:

Elasticidade Negativa (quando aumentos de preo levam  reduo da quantidade)

Elasticidade Unitria (quando alteraes de preo resultam em variaes proporcionais na quantidade)

Alta Elasticidade (quando pequenas variaes de preo causam grandes variaes na quantidade)

Elasticidade Zero (quando a quantidade permanece inalterada, mesmo com mudanas no preo)

Aqui esto estratgias de precificao concisas para cada segmento identificado no grfico:

Elasticidade Negativa: Foque em reduzir os preos para estimular a demanda, j que os consumidores so sensveis a aumentos de preo. Cortes nos preos podem gerar aumentos significativos na quantidade vendida, compensando as margens mais baixas.

Elasticidade Unitria: Mantenha os preos estveis e concentre-se em melhorar o valor percebido do produto ou diferenciar as ofertas. Como as mudanas de preo resultam em variaes proporcionais na quantidade, o foco deve ser equilibrar o preo com o valor percebido para manter o volume de vendas.

Alta Elasticidade: Utilize estratgias de precificao dinmica, como descontos promocionais ou redues de preo em perodos de alta demanda, para aproveitar a alta sensibilidade dos consumidores s mudanas de preo. Pequenas redues de preo podem levar a grandes aumentos no volume de vendas.

Elasticidade Zero: Ajustes de preo so menos eficazes para impulsionar vendas nesse segmento. O foco deve estar em estratgias no relacionadas ao preo, como aprimoramento de funcionalidades, qualidade do produto ou atendimento ao cliente, para se diferenciar e ganhar participao de mercado sem depender de alteraes de preo.

A Elasticidade-Preo da Demanda (PED) mede a sensibilidade da quantidade demandada de um produto em relao s variaes em seu preo. Ela  calculada como a variao percentual na quantidade demandada dividida pela variao percentual no preo. Essa mtrica ajuda as empresas a entender o quo sensveis seus clientes so a mudanas de preo, permitindo que tomem decises de precificao mais informadas.

##Anlise com mtodo midpoint

A Elasticidade-Preo da Demanda pelo Ponto Mdio (tambm chamada de Elasticidade-Arco)  uma forma mais precisa de calcular a elasticidade ao longo de um intervalo de preos e quantidades  em vez de usar apenas os valores iniciais.

```python
df = pd.read_csv("/content/Competition_Data.csv")
```

```python
df['Fiscal_Week_ID']  = df['Fiscal_Week_ID'].astype('string')
df['Store_ID'] = df['Store_ID'].astype('string')
df['Item_ID'] = df['Item_ID'].astype('string')
```

```python
df.head()
```

```python
# Calcular diferenas e mdias mveis
df['Delta_Q'] = df['Item_Quantity'] - df['Item_Quantity'].shift(1)
df['Delta_P'] = df['Price'] - df['Price'].shift(1)
df['Avg_Q'] = (df['Item_Quantity'] + df['Item_Quantity'].shift(1)) / 2
df['Avg_P'] = (df['Price'] + df['Price'].shift(1)) / 2

df['res_qty'] = df['Delta_Q'] / df['Avg_Q']
df['res_prc'] = df['Delta_P'] / df['Avg_P']

# Calcular Elasticidade com mtodo do ponto mdio
df['PED_mid'] = df['res_qty'] / df['res_prc']

# Tratar valores infinitos e nulos
df.replace([float('inf'), -float('inf')], np.nan, inplace=True)
df.dropna(subset=['PED_mid'], inplace=True)

# Funo de resumo estatstico
def resumo_estatistico(df, coluna):
    resumo = df[coluna].describe(percentiles=[0.25, 0.5, 0.75])
    print(resumo)

# Exibir estatsticas da elasticidade
resumo_estatistico(df, 'PED_mid')
```

```python
#definindo os threshholds de PED
elastic_threshold = 1
inelastic_threshold = -1

# segment the data based on PED
df['Segment'] = 'Unitary Elastic'
df.loc[df['PED_mid'] > elastic_threshold, 'Segment'] = 'Highly Elastic'
df.loc[df['PED_mid'] < inelastic_threshold, 'Segment'] = 'Inelastic'
df.loc[df['PED_mid'] == 0, 'Segment'] = 'Zero Elasticity'
df.loc[df['PED_mid'] < 0, 'Segment'] = 'Negative Elasticity'

# count the number of items in each segment
segment_counts = df['Segment'].value_counts()

segment_counts
```

```python
fig_segment = px.scatter(df, x='res_prc', y='res_qty', color='Segment',
                         title='Price Change vs Quantity Change Across Different Market Segments',
                         labels={'res_prc': 'Price Change (%)', 'res_qty': 'Quantity Change (%)'},
                         color_continuous_scale='Viridis', opacity=0.6)

fig_segment.update_layout(xaxis_title='Price Change (%)', yaxis_title='Quantity Change (%)',
                          template='plotly_white')

fig_segment.show()
```

Eixos:

Eixo X: Variao percentual no preo (% P)

Eixo Y: Variao percentual na quantidade (% Q)

Cores (Segmentos):

 Negative Elasticity: P  leva Q   comportamento normal de bens tpicos

 Highly Elastic: pequenas mudanas em preo  grandes mudanas em quantidade (|PED| > 1)

 Zero Elasticity: variao de preo sem impacto na quantidade (PED  0)

Concluses com base no mtodo do ponto mdio
Segmentao clara:
O uso do midpoint permitiu uma segmentao robusta, pois normaliza as variaes usando mdias dos pontos (mais estvel do que base inicial).
